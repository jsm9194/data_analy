## 인덱스 설정 및 리셋

```python
# 데이터 프레임 왼쪽 진하게 처리되어있는게 인덱스임
df.index

# 0부터 5까지 하나씩 올라가는 인덱스임
```

```python
# 데이터 프레임의 인덱스는 df.index를 명시하여 접근할 수 있음
df.index = ['one', 'two', 'three', 'four', 'five'] 
# 인덱스의 값을 바꾸고 싶다면 인덱스의 개수만큼의 원소를 포함한 리스트를 할당하여 
df
```

```python
# 새로운 컬럼 성별(sex) 생성
sex = pd.Series(['여자', '여자', '남자', '남자', '남자'])
df['성별'] = sex # 성별 컬럼에 새롭게 만든 시리즈를 값으로 부여
df

#결측치 nan으로 나옴
```

```python
# 성별에 대한 인덱스 확인
sex.index
```

```python
# 시리즈 sex 와 데이터 프레임 df의 인덱스 비교해보자
print('sex의 인덱스 > ' , sex.index )
print('df의 인덱스 > ' , df.index )
```

### 인덱스를 아까 글자로 바꿔놔서 매칭이 안됐구나!!

```python
# 인덱스 통일후 다시 성별 컬럼에 값 부여 
sex.index = df.index # 성별 시리즈의 인덱스르르 데이터 프레임 인덱스로 변경
df['성별'] = sex
df
```

```python
# 현재 설정되어 있는 인덱스 리셋
df.reset_index(inplace=True) # inplace=True 값을 괄호안에 넣어주면 원본에 바로 적용되어버려
#df.reset_index()
```

```python
df
```

```python
# index 설정
df.set_index('index', inplace=True)
#df.set_index
```

## 컬럼삭제

```python
# 성별 컬럼 삭제(dataframe의 drop매서드 이용)
df.drop('성별', axis=1) # drop에서 axis=1의 값을 주면 컬럼을 삭제하겠다.
# inplace인수를 안줘서 원본에 반영은 안된것
```

```python
df.drop('성별', axis='columns') # 컬럼즈로 직접 명시해줘도 된다
```

```python
# 컬럼을 완전히 삭제하는 방법

#방법1 새로운 데이터 프레임 생성
df2 = df.drop('성별', axis=1)
df2
```

```python
# 원본데이터 확인
df
```

```python
# 방법2 inplace =True
df.drop('성별', axis=1, inplace=True)
df
```

```python
# 데이터 프레임 행 삭제
df.drop('one', axis=0) # one라는 인덱스를 가진 행을 삭제 axis0
```
