## dataFrame Merge

```python
#판다스 패키지 로딩
import pandas as pd
```

## Inner Join

```python
# 데이터 프레임 생성

data1 = {
    '학번':[1, 2, 3, 4],
    '이름':['아이유', '김연아', '홍길동', '강감찬'],
    '학과':['철학', '경영학', '컴퓨터', '물리학']
}

data2 = {
    '학번':[1, 2, 4, 5],
    '학년':[2, 4, 3, 1],
    '학점':[1.5, 2.0, 4.1, 3.8]
}

df1 =pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
# inner join -일치되는 것만 merge
inner_df = pd.merge(df1, df2, on='학번', how='inner') 
#어떤 컬럼을 기준으로 merge할지 on파라메터에 명시한다
```

```python
display(inner_df)
```

## outer Join

```python
# 위와 같은 데이터 사용

outer_df = pd.merge(df1, df2, on='학번', how='outer') 
```

```python
display(outer_df)
```

## Left and Right Join

```python
left_df = pd.merge(df1, df2, on='학번', how='left')
# pd.merge(첫번째 데이터프레임, 두번째 데이터 프레임, on, how)

display(left_df)# df1의 키를 기준으로 merge 진행됨 학번5가 없음
display(df1)
```

```python
#right join 오른쪽거 기준으로 하겠다 학번3이 없음
right_df = pd.merge(df1, df2, on='학번', how='right')
display(right_df)
```

## merge 응용 : 현실데이터를 다룰때 생길만한거

### 1)같은 의미의 다른 컬럼명을 병합할때

```python
# case 1: 같은의미의 다른 컬럼명 > 학번과 학생 고유번호

data1 = {
    '학번':[1, 2, 3, 4],
    '이름':['아이유', '김연아', '홍길동', '강감찬'],
    '학과':['철학', '경영학', '컴퓨터', '물리학']
}

data2 = {
    '학생고유번호':[1, 2, 4, 5],
    '학년':[2, 4, 3, 1],
    '학점':[1.5, 2.0, 4.1, 3.8]
}

df1 =pd.DataFrame(data1)
df2 = pd.DataFrame(data2)


display(df1)
display(df2)
```

```python
merged_df = pd.merge(df1, df2, left_on='학번', right_on='학생고유번호', how='inner') 
# pd.merge(첫번째 데이터프레임, 두번째 데이터프레임,left_on, right_on, how='원하는 방식')

display(merged_df)
# 서로다른값으로 보고 그냥 합쳐졌음
```

```python
# 학번과 학생고유번호중 하나를 삭제 drop메서드 사용
#merged_df.drop('학생고유번호', axis=1)

merged_df.drop('학생고유번호', axis=1, inplace=True)
display(merged_df)
```

### 공유하는 컬럼이 없는 경우

```python
data1 = {
    '학번':[1, 2, 3, 4],
    '이름':['아이유', '김연아', '홍길동', '강감찬'],
    '학과':['철학', '경영학', '컴퓨터', '물리학']
}

#학번 제거했음
data2 = {
    '학년':[2, 4, 3, 1],
    '학점':[1.5, 2.0, 4.1, 3.8]
}

df1 =pd.DataFrame(data1)
df2 = pd.DataFrame(data2, index=[1, 2, 4, 5]) #예제를 위한 이상한 인덱스 부여


display(df1)
display(df2)
```

```python
# 양 데이터 프레임의 인덱스를 기준으로 병합 진행
merged_df1 = pd.merge(df1, df2, left_index=True, right_index=True, how='inner')
```

```python
display(merged_df1)
```

```python
# 왼쪽은 학번을 기준으로
# 오른쪽은 인덱스를 기준으로 병합
merged_df2 = pd.merge(df1, df2, left_on='학번', right_index=True, how='inner')
display(merged_df2)
```

```python
# join 사용 양쪽 데이터 프레임의 인덱스를 기준으로 병함
df1.join(df2, how='inner') # index기준
```

## 데이터 연결 pandas concat

```python
import numpy as np
import pandas as pd

np.random.seed(1) #예제를 위해 넘파이 활용 고정 난수 설정
```

```python
# 3x2 정수로 구성된 매트릭스 생성
df1 = pd.DataFrame(np.random.randint(0, 9, (3,2)), # 3x2  매트릭스
                   # random integer 만드는데 0과 9사이 3행 2열 
                  index=['a', 'b', 'c'],
                    columns=['one', 'two'])

display(df1)
```

```python
df2 = pd.DataFrame(np.random.randint(0, 9, (2,2)), # 2x2  매트릭스
                   # random integer 만드는데 0과 9사이 3행 2열 
                  index=['a', 'b'],
                    columns=['three', 'four'])
display(df2)
```

```python
# df1과 df2를 concat으로 연결

result_df = pd.concat([df1, df2], axis=1) # 리스트로 데이터 프레임을 주고 열방향연결
display(result_df)
```

```python
result_df2 = pd.concat([df1, df2], axis=0) # 리스트로 데이터 프레임을 주고 행방향연결
display(result_df2)
```

```python
#outer join으로도 concat과 같은 결과 얻을수 있음
merge_df= pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
display(merge_df)
```
